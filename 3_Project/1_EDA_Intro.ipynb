```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
df_DA_DE = df[(df['job_country'] == 'Germany') & (df['job_title_short'] == 'Data Analyst')]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data
df_DE_plot = df_DA_DE['job_location'].value_counts().head(10).to_frame()
df_US_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()

# Create the subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Plot for Germany
sns.barplot(data=df_DE_plot, x='count', y='job_location', hue='count', palette='Blues', ax=axes[0])
axes[0].set_title('Counts of Job Locations for Data Analyst in Germany')
axes[0].set_xlabel('Number of Jobs')
axes[0].set_ylabel('')
axes[0].legend().remove()

# Plot for Cyprus
sns.barplot(data=df_US_plot, x='count', y='job_location', hue='count', palette='Oranges', ax=axes[1])
axes[1].set_title('Counts of Job Locations for Data Analyst in Cyprus')
axes[1].set_xlabel('Number of Jobs')
axes[1].set_ylabel('')
axes[1].legend().remove()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
df_plot = df['job_country'].value_counts().to_frame().head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df['company_name'].value_counts().to_frame()[1:].head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Company')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

us_colors = sns.color_palette('Oranges', 2)
de_colors = sns.color_palette('Blues', 2)

for i, (column, title) in enumerate(dict_column.items()):
    axes[0, i].pie(
        df_DA_US[column].value_counts(), 
        labels=['False', 'True'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=us_colors
    )
    axes[0, i].set_title(f'{title} (US)')

for i, (column, title) in enumerate(dict_column.items()):
    axes[1, i].pie(
        df_DA_DE[column].value_counts(), 
        labels=['False', 'True'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=de_colors
    )
    axes[1, i].set_title(f'{title} (DE)')


plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data for US and DE
df_US_plot = df_DA_US['company_name'].value_counts().head(10).reset_index()
df_US_plot.columns = ['company_name', 'count']

df_DE_plot = df_DA_DE['company_name'].value_counts().head(10).reset_index()
df_DE_plot.columns = ['company_name', 'count']

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Plot for the US
sns.barplot(data=df_US_plot, x='count', y='company_name', palette='Oranges', ax=axes[0])
axes[0].set_title('Top 10 Companies for Data Analysts (US)')
axes[0].set_xlabel('Number of Jobs')
axes[0].set_ylabel('Company Name')
axes[0].grid(axis='x', linestyle='--', alpha=0.7)

# Plot for Germany
sns.barplot(data=df_DE_plot, x='count', y='company_name', palette='Blues', ax=axes[1])
axes[1].set_title('Top 10 Companies for Data Analysts (DE)')
axes[1].set_xlabel('Number of Jobs')
axes[1].set_ylabel('Company Name')
axes[1].grid(axis='x', linestyle='--', alpha=0.7)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
