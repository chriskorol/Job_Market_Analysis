```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df.set_index('job_posted_date', inplace=True) # This is code from the last problem
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
print(df.columns)
```

```python
df_pivot_table = df.pivot_table(index='company_name', values='salary_year_avg', aggfunc='median').sort_index(ascending=False)
df_pivot_table.head()
```

```python
df_pivot_table = df.pivot_table(index='job_title_short',values='salary_year_avg',aggfunc='median')
df_pivot_table.reset_index(inplace=True)
df_pivot_table.set_index('salary_year_avg', inplace=True)
df_pivot_table.head()
```
